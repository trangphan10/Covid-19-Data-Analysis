```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis = 1)
```

```python
covid_df.head(10)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head(10)
```

```python
# Active cases 
covid_df['Active cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.head()
```

```python
statewise = pd.pivot_table(covid_df, values=['Confirmed','Cured','Deaths'],index = 'State/UnionTerritory')
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by='Confirmed',ascending= False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
# Top 10 active cases states 
Top_10_active_state = covid_df.groupby(by='State/UnionTerritory').max()[['Active cases','Date']].sort_values(by = ['Active cases'],ascending=False).reset_index()
figure = plt.figure(figsize=(16,9))
ax = sns.barplot(data = Top_10_active_state.iloc[:10],x = 'State/UnionTerritory',y='Active cases',linewidth=2,edgecolor='red')
plt.xlabel('States') 
plt.ylabel('Total active cases')
plt.show()
```

```python
# Top 10 states with highest death  
Top_10_states_death = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = 'Deaths',ascending = False).reset_index()
fig = plt.figure(figsize=(16,9)) 
ax = sns.barplot(data = Top_10_states_death.iloc[:12],x = 'State/UnionTerritory',y='Deaths',linewidth=2,edgecolor='yellow')
plt.xlabel('States')
plt.ylabel('Total Death cases') 
plt.title('Top 10 states with highest death in India')
plt.show()
```

```python
# growth trend 
figure = plt.figure(figsize=(16,8)) 
ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Tamil Nadu','Delhi','Uttar Pradesh'])],
                  x = 'Date',y='Active cases', hue = 'State/UnionTerritory')
ax.set_title('Top 5 Affected States in India',size =16)
```

```python
# Drop rows that State is India 
vaccine = vaccine_df[vaccine_df.State != 'India']
```

```python
vaccine.head(5)
```

```python
vaccine.rename(columns={'Total Doses Administered': 'Total'},inplace= True)
vaccine.head(5)
```

```python
# Find any states that has the most confirmed vaccine people and the least confirmed vaccine people
Top_states_with_most_vaccine = vaccine.groupby(by='State')['Total'].sum().to_frame('Total').sort_values(by='Total',ascending=False).reset_index()
fig = plt.figure(figsize=(16,16)) 
ax = sns.barplot(data=Top_states_with_most_vaccine.iloc[:10],x = 'State',y='Total',linewidth=2,edgecolor='red')
plt.xlabel('State')
plt.ylabel('Total Individuals Vaccination')
plt.title('Top 10 states with most confirmed vaccination people')
plt.show()
```
